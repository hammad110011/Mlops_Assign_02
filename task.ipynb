```python
from bs4 import BeautifulSoup
import requests

# Function to fetch links, titles, and descriptions from a webpage
def fetch_web_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extracting links, titles, and descriptions
    link_list = []
    title_list = []
    description_list = []

    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('http'):
            link_list.append(href)
            title = link.get_text(strip=True)
            title_list.append(title)
            # Extracting description if available
            description = link.find_next_sibling(string=True)
            if description and len(description.strip()) > 0:
                description_list.append(description.strip())
            else:
                description_list.append("")

    return link_list, title_list, description_list

# Fetch data for BBC News
BBC_News_Links, BBC_News_Titles, BBC_News_Descriptions = fetch_web_data('https://www.bbc.com/')

# Print the extracted data for BBC News
print("\nBBC News:")
for link, title, description in zip(BBC_News_Links, BBC_News_Titles, BBC_News_Descriptions):
    print(f"Link: {link}\nTitle: {title}\nDescription: {description}\n")

# Fetch data for Dawn News
Dawn_News_Links, Dawn_News_Titles, Dawn_News_Descriptions = fetch_web_data('https://www.dawn.com/')

# Print the extracted data for Dawn News
print("Dawn News:")
for link, title, description in zip(Dawn_News_Links, Dawn_News_Titles, Dawn_News_Descriptions):
    print(f"Link: {link}\nTitle: {title}\nDescription: {description}\n")


```
